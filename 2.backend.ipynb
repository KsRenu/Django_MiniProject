```python
# functions involved
def sepDeptYear(dept,df_final):
  nstud=0
  yr=int(input('Enter the year of this department students (2 or 3): '))
  for i, row in df.iterrows():
    if df.loc[i,'department']==dept:
      if df.loc[i,'rollNo'][:2]=='20':
                data=df.iloc[i].tolist()
                df_2=pd.DataFrame(columns=df.columns.tolist())
                df_2.loc[len(df_2.index)] = data
                if yr==2:
                  nstud+=1
                  df_final.loc[len(df_final.index)] = data
      if df.loc[i,'rollNo'][:2]=='19':
                data=df.iloc[i].tolist()
                df_3=pd.DataFrame(columns=df.columns.tolist())
                df_3.loc[len(df_3.index)] = data
                if yr==3:
                  nstud+=1
                  df_final.loc[len(df_final.index)] = data
  name2='d_'+ dept + '_2.csv'
  name3='d_'+ dept + '_3.csv'
  df_2.to_csv(name2, index = False)
  df_3.to_csv(name3, index = False)
  df_final.to_csv(r'df_final.csv',index=False)
  df_2 = df_2.drop(labels=range(0,len(df_2.index)-1), axis=0)
  df_3 = df_3.drop(labels=range(0,len(df_3.index)-1), axis=0)
  return nstud




def allocate(l):
  i=1
  xyz=36
  for x in range(1,nhall+1):
    print('Hall : ',x)
    while(True):
      if i<len(l)-1:
        if i%6==0:
          print(x,'_',i,'_',l[i+1],'      ')
          print('\n')
        else:
          print(x,'_',i,'_',l[i+1],'      ',end='')
        if i==xyz:
          xyz+=xyz
          break
      else:
        break
      
      i+=1
    print('-----------------------------')
```
